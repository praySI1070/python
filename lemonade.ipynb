```python
import tensorflow as tf
import pandas as pd
```

```python
#파일로부터 데이터 읽어 오기
file_adress = 'https://raw.githubusercontent.com/praySI1070/python/main/polyclass/lemonade%20(1).csv'
lemonade = pd.read_csv(file_adress)
```

```python
print(lemonade.shape)
```

```python
print(lemonade.columns)
```

```python
independent = lemonade[['온도']]
dependent = lemonade[['판매량']]
print(dependent.shape,independent.shape)
```

```python
lemonade.head()
```

```python
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```python
model.fit(independent, dependent, epochs=1000)
```

```python
print("Predictions :", model.predict([[15]]))
```

```python
model.fit(independent, dependent, epochs=10)
```
